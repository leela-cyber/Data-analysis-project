Analyze the provided company dataset `registered_companies.csv` by first clustering companies based on their 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' to categorize their financial status (financially weak, moderately funded, financially strong), then classify companies into 'Compliant' or 'Non-Compliant' based on their latest annual return and financial statement filing status, and finally summarize the insights from both analyses.

## Prepare Data for Clustering

### Subtask:
Select the 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' columns for clustering. Handle any potential outliers and scale these numerical features to ensure they contribute equally to the clustering process.


**Reasoning**:
The first instruction is to select the specified columns 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' and store them in a new DataFrame for clustering.



````python
df_clustering = df[['AUTHORIZED_CAP', 'PAIDUP_CAPITAL']].copy()
print("Selected columns for clustering:")
display(df_clustering.head())
````

**Reasoning**:
The next step is to identify and handle outliers in the 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' columns using the Interquartile Range (IQR) method as instructed. This involves calculating Q1, Q3, IQR, and then capping the outliers for each column.



# Task
```python
for column in ['AUTHORIZED_CAP', 'PAIDUP_CAPITAL']:
    Q1 = df_clustering[column].quantile(0.25)
    Q3 = df_clustering[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df_clustering[column] = df_clustering[column].clip(lower=lower_bound, upper=upper_bound)

print("Outliers handled for 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' columns using IQR capping.")
print("DataFrame after outlier handling:")
display(df_clustering.describe())
```

## Prepare Data for Clustering

### Subtask:
Handle outliers in the 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' columns of `df_clustering` using the IQR method.


**Reasoning**:
The next step is to handle outliers in the 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' columns of the `df_clustering` DataFrame using the IQR method, as instructed. This involves calculating Q1, Q3, IQR, and then capping the outliers for each column, followed by displaying descriptive statistics to observe the effect.



````python
for column in ['AUTHORIZED_CAP', 'PAIDUP_CAPITAL']:
    Q1 = df_clustering[column].quantile(0.25)
    Q3 = df_clustering[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_clustering[column] = df_clustering[column].clip(lower=lower_bound, upper=upper_bound)

print("Outliers in 'AUTHORIZED_CAP' and 'PAIDUP_CAPITAL' columns have been handled using the IQR method.")
print("Descriptive statistics of the DataFrame after outlier capping:")
display(df_clustering.describe())
````
